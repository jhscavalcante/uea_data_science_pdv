## Sobre o Dataset

O dataset a ser utilizado é o *IBM HR Analytics Employee Attrition & Performance*, disponível no Kaggle. O dataset contém informações sobre funcionários de uma empresa fictícia, como idade, gênero, estado civil, nível de satisfação, entre outros.

[Link do Kaggle](https://www.kaggle.com/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset/data)

O termo *attrition* ("atrito") é frequentemente usado de forma intercambiável no contexto de recursos humanos e gestão de força de trabalho, mas ele se refere a conceitos ligeiramente diferentes em relação às mudanças no número e na composição dos empregados dentro de uma organização.

# Respondendo aos questionamentos

## 1. Pergunta: 
### Qual é o percentual e quantidade de funcionários que permaneceram e saíram da empresa?

```python
# Mapeia os valores para rótulos mais claros, sem alterar o df original
mapa_valores = {
    'No': 'Permanecem na empresa',
    'Yes': 'Saíram da empresa'
}
counts = df['Attrition'].value_counts().rename(index=mapa_valores)
total = counts.sum()

# Cria o gráfico
ax = counts.plot(kind='bar', figsize=(10, 7), rot=0, legend=False, color=["#3497C4", "#FF6961"])

# Itera sobre as barras para adicionar os rótulos
for patch in ax.patches:
    # A altura da barra é a contagem absoluta
    height = int(patch.get_height())
    
    # Calcula o percentual
    percentage = (height / total) * 100
        
    # Cria um rótulo de múltiplas linhas com o percentual e a contagem
    label_text = f"{percentage:.1f}%\n({height:,})".replace(',', '.')
    
    # Posição X (centro da barra)
    x = patch.get_x() + patch.get_width() / 2
    
    # Posição Y (agora centralizada na barra para acomodar duas linhas)
    y = height / 2
    
    # Adiciona o texto, agora com espaçamento entre as linhas
    ax.text(x, y, label_text, 
            ha='center', 
            va='center', 
            color='white', 
            fontweight='bold',
            fontsize=12,
            linespacing=1.5) # Parâmetro para melhorar a leitura de múltiplas linhas

# Títulos e rótulos
plt.title("Funcionários por Status", fontsize=16)
plt.xlabel("Status", fontsize=12)
plt.ylabel("Quantidade de Funcionários", fontsize=12)

# Ajustes estéticos
plt.ylim(0, counts.max() * 1.1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#plt.tight_layout()

plt.show()
```

## 2. Pergunta: 
### Das pessoas que permanecem na empresa, quais são os cargos com mais funcionários?

```python
# restrigindo para os funcionários que permanceram na empresa
df_com_attrition = df[df['Attrition'] == 'Yes']

# obtendo o total por cargo
contagem_cargos = df_com_attrition['JobRole'].value_counts()

# calculando o percentual
total_funcionarios = contagem_cargos.sum()
percentual_cargos = (contagem_cargos / total_funcionarios * 100).sort_values(ascending=False)

# configurando o top 3
n_top = 3
cor_destaque = '#4A6A9E'  # Azul mais escuro
cor_resto = '#DCDCDC'     # Cinza claro

cores = [cor_destaque if i < n_top else cor_resto for i in range(len(percentual_cargos))]

fig, ax = plt.subplots(figsize=(12, 8))

bars = ax.barh(percentual_cargos.index, percentual_cargos.values, color=cores)

for bar in bars:
    width = bar.get_width()
    label = f'{width:.2f}%'
    ax.text(width + 0.5, bar.get_y() + bar.get_height()/2, label,
    va='center', ha='left', fontsize=10)

top_cargos = percentual_cargos.index[:n_top].tolist()
top_soma = percentual_cargos.iloc[:n_top].sum()
titulo = f'Os 3 cargos representam mais de {top_soma:.0f}% do total de funcionários que permanecem na empresa' 
ax.set_title(titulo, fontsize=14, pad=20)

ax.get_xaxis().set_visible(False)
ax.spines[['top', 'right', 'bottom', 'left']].set_visible(False)
ax.tick_params(axis='y', length=0, labelsize=11)

ax.invert_yaxis()

plt.tight_layout()

plt.show()

```

## 3. Pergunta: 
### Como está a distribuição de Idade por gênero e status?

### Análise prévia

```python

df.groupby(["Gender", "Attrition"], observed=False)["Age"].describe()
```

### Gráfico

```python
df_para_plot = df.copy()

# Dicionários para as traduções
mapa_genero = {'Female': 'Feminino', 'Male': 'Masculino'}
mapa_attrition = {'No': 'Não saiu da empresa', 'Yes': 'Saiu da empresa'}

# Aplicando as traduções na cópia do DataFrame
df_para_plot['Gender'] = df_para_plot['Gender'].map(mapa_genero)
df_para_plot['Attrition'] = df_para_plot['Attrition'].map(mapa_attrition)

# Capturamos o eixo (ax) para customizar os títulos depois.
ax = df_para_plot.boxplot(
    column="Age", 
    by=["Gender", "Attrition"], 
    grid=False, 
    showmeans=True,
    figsize=(10, 7)
)

# Remove o título principal automático gerado pelo pandas
plt.suptitle('')

# Define novos títulos e rótulos em português
ax.set_title('Distribuição de Idade por (Gênero e Status)', fontsize=14)
ax.set_ylabel('Idade')
ax.set_xlabel('')

# Melhora a legibilidade dos rótulos do eixo X se necessário
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

## 4. Pergunta: 
### Qual o total de funcionários que saíram ou não da empresa por tempo de casa?

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

df_plot = df.copy()
df_plot['Status'] = df_plot['Attrition'].map({'No': 'Não saiu da empresa', 'Yes': 'Saiu da empresa'})

# Agrupamos os dados para contar o total de pessoas por ano na empresa e por status (saiu/não saiu)
saida_por_ano = df_plot.groupby(['YearsAtCompany', 'Status']).size().reset_index(name='Total')

# Definimos o tamanho do gráfico
plt.figure(figsize=(14, 8))

# Criamos o gráfico de linhas usando a biblioteca seaborn, que lida com a legenda automaticamente
ax = sns.lineplot(
    data=saida_por_ano,
    x='YearsAtCompany',
    y='Total',
    hue='Status',
    marker='o', # Adiciona um marcador de círculo em cada ponto de dados
    linewidth=2.5 # Aumenta a espessura da linha
)

# Adicionamos o título e os rótulos dos eixos, conforme a imagem de referência
plt.title('Total de funcionários que saíram ou não da empresa por tempo de casa', fontsize=16, weight='bold')
plt.xlabel('Anos na Empresa', fontsize=12)
plt.ylabel('Total de Funcionários', fontsize=12)

# Melhorando os marcadores do eixo X para exibir os anos de forma mais clara
plt.xticks(ticks=range(0, saida_por_ano['YearsAtCompany'].max() + 2, 2))
plt.xlim(0, saida_por_ano['YearsAtCompany'].max() + 1) # Define o limite do eixo x para começar em 0

# Exibindo o gráfico
plt.show()
```
